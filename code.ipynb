```python
from pycaret.classification import *
import pandas as pd
import mlflow
from mlflow.models import infer_signature

# Set URI for MLflow tracking server
mlflow.set_tracking_uri(uri="http://127.0.0.1:7000")

# Set experiment name
mlflow.set_experiment("Mlflow w9")

# Import data
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

# Filter data to include only the desired features
desired_features = ['Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep',
                   'Physical Activity Level', 'Stress Level', 'BMI Category', 'Heart Rate', 'Daily Steps']
data = data[desired_features + ['Sleep Disorder']]

# Initialize PyCaret setup with desired features
s = setup(data, target='Sleep Disorder', 
        #   ignore_features=['Gender', 'Age', 'Occupation'],  # Ignoring the non-desired features
          bin_numeric_features=['Gender', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'BMI Category'])

# pca=True, pca_components=10
# Compare models
best = compare_models()

# Tune the best performing model
tuned_model = tune_model(best)

# Evaluate the model
evaluate_model(tuned_model)

# Finalize the model for deployment
final_model = finalize_model(tuned_model)

# Save the finalized model
save_model(final_model, 'sleep_disorder_classification_model')

# Enable autologging
mlflow.autolog()

# Start a new MLflow run
with mlflow.start_run():
    # Log the trained model
    mlflow.sklearn.log_model(final_model, "sleep_disorder_classification_model")
```
